```python
import pandas as pd
import numpy as np
```

```python
df_2022 = pd.read_parquet("../data/2022_customer_data.parquet")
df_2022.describe()
```

```python
df_2022.head()
```

We'll apply the same cleaning steps as we did for the previous data

```python
df_2022 = df_2022.sort_values(by="last_purchase_date").reset_index(drop=True)


df_2022["first_purchase_date"] = pd.to_datetime(df_2022["first_purchase_date"].dt.strftime("%Y-%m-%d %H:%M"))
df_2022["last_purchase_date"] = pd.to_datetime(df_2022["last_purchase_date"].dt.strftime("%Y-%m-%d %H:%M"))

# confirms avg_order_value calculation
df_2022["avg_order_value"] = (df_2022["total_spent"] / df_2022["transaction_count"]).round(3)


df_2022.rename(columns={"customer_age": "customer_tenure"}, inplace=True)
df_2022.rename(columns={"promo_code_used": "promo_usage_count"}, inplace=True)


CONVERSION_RATE = 16787.54   # This is the rate as of January 30 2026, can change ofc

df_2022["total_spent"] = (df_2022["total_spent"] / CONVERSION_RATE).round(3)
df_2022["avg_order_value"] = (df_2022["avg_order_value"] / CONVERSION_RATE).round(3)
```
